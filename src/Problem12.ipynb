## Train with Different GLoVE embedding dimensions

```python
accuracies=[]
dimensions=[50,100,200,300]

for dimension in dimensions:
    test_acc,_,_,_=pipeline(df,normalize="mean",glove_embedding_dims=dimension)
    accuracies.append(test_acc)
```

```python
plt.plot(dimensions,accuracies)
plt.xlabel("Dimensions of Glove Embeddings")
plt.ylabel("Accuracy")
plt.title("Accuracy vs Dimensions of Glove Embeddings")
plt.savefig("Problem11.png")
plt.show()
```
